```python
# Read the Excel file
df = pd.read_excel('/content/tableemail.xlsx', header = 2)

# Email template
email_template = """
Dear {name},

I hope this email finds you well.
I have generated this mail using Python.
Included your mail id for testing.
Please ignore

Warm regards,
Piyush Deshpande.
"""
```

```python
# Sender email credentials
sender_email = 'pxxxxxxx@gmail.com'
sender_password = 'xxxx xxxx xxxx xxxx'

# SMTP server configuration (for Gmail)
smtp_server = 'smtp.gmail.com'
smtp_port = 587

# Establish connection to SMTP server
server = smtplib.SMTP(smtp_server, smtp_port)
server.starttls()
server.login(sender_email, sender_password)

```

```python
# Iterate through each row in the Excel file
for index, row in df.iterrows():
    # Extract name and email address
    name = row['Name']
    email = row['Email']

    # Personalize the email template
    personalized_email = email_template.format(name=name)

    # Create message container
    msg = MIMEMultipart()
    msg['From'] = sender_email
    msg['To'] = email
    msg['Subject'] = 'Interest in Brain-Related Research Projects and Request for Financial Aid'

    # Attach email body
    msg.attach(MIMEText(personalized_email, 'plain'))
    # Attach files (sample_attachment1.txt, sample_attachment2.txt, sample_attachment3.txt are example file names)
    attachment_paths = ['/content/Best paper Certificate (Literature).pdf',
                        '/content/Comparative Study for Licence Plate Detection.pdf',
                        '/content/Electromyography based System for Human Postures.pdf',
                        '/content/Licence Plate Detection and Recognition.pdf',
                        '/content/PIYUSH_DESHPANDE_ACADEMIC_TRANSCRIPT.PDF',
                        '/content/Piyush_Deshpande_Resume.pdf',
                        '/content/Staircase Detection for Visually Impaired.PDF'
                        ]
    for attachment_path in attachment_paths:
        attachment_name = attachment_path[9:]
        with open(attachment_path, 'rb') as attachment:
            part = MIMEBase('application', 'octet-stream')
            part.set_payload(attachment.read())
        encoders.encode_base64(part)
        part.add_header('Content-Disposition', f'attachment; filename= {attachment_name}')
        msg.attach(part)

    # Send email
    server.send_message(msg)

    # Print confirmation
    print(f"Email sent to {name} at {email}")

# Close the connection to the SMTP server
server.quit()
```
